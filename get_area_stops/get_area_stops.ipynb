Instead of looking at the dataset for all of England, Scotland and Wales, we
might just be interested in stops within an area. Using the
[ATCO area codes](using_atco_codes.ipynb) we can limit the API request to a
specific set of stops — meaning less post-filtering and a faster response time.

For example, to get just the stops of Leicestershire we can pass its area code
of '260' into the `naptan.get_area_stops` function.

<div class="admonition info">
<p class="admonition-title">Info</p>
<p>This function requires a list be passed into the <code>area_codes</code> parameter. Even just for a single area.<br>The code can be passed in as a number or a string.</p>
</div>

```python
import naptan

# pass in the area code for Leicestershire
naptan.get_area_stops(['260']) # passed in as a list
```

---

We can ask for an arbitrary number of areas at once. For example, we can get both
Leicester *and* Leicestershire by passing in their respective area codes (269, 260).

```python
import naptan

naptan.get_area_stops([269, 260])
```

---

If we want to do larger areas we can let the `naptan.ATCO_CODES` table do some of
the heavy lifting. For example, to get the stops for all of the East Midlands
region we can query the lookup table for the "EM" region, and pass this to the
function.

```python
import naptan

east_mids_codes = naptan.ATCO_CODES.query('Region == "EM"')['Code']

naptan.get_area_stops(east_mids_codes)
```
