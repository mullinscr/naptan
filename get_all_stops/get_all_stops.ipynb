To get all the stops from the NaPTAN dataset, we use the `naptan.get_all_stops`
function.

This will return the data as a pandas.DataFrame.

```python
import naptan

naptan.get_all_stops()
```

By default the returned dataset includes stops of every "status" (can be one of
'active', 'inactive', or 'pending'). To return just the 'active' stops we pass
in `status='active'` to the function.

```python
import naptan

naptan.get_all_stops(status='active')
```
